```python
# Function to convert age values to numerical categories
def convert_age(age):
    # Check for missing values (NaN, empty string, etc.)
    if pd.isna(age) or age == '' or age is None:
        return 2  # missing = 2
    
    # Convert to string and clean
    age_str = str(age).strip().lower()
    
    # Try to extract numerical value
    try:
        import re
        numbers = re.findall(r'\d+', age_str)
        if numbers:
            age_num = int(numbers[0])
            
            # Apply the rules
            if age_num < 25:
                return 1
            elif 25 <= age_num <= 29:
                return 2
            elif age_num >= 30:
                return 3
            else:
                return 2  # anything else = 2
        else:
            return 2  # anything else = 2
    except:
        return 2  # anything else = 2

# Directly replace the A1 column
df['A1'] = df['A1'].apply(convert_age)

# Verify the replacement
print("Unique values in A1 after replacement:")
print(df['A1'].value_counts().sort_index())
print("\nValue counts:")
print(df['A1'].value_counts())

# Check first few rows
print("\nFirst 10 rows of A1 after replacement:")
print(df['A1'].head(10))
```

```python
# Convert A2 column (gender) to numerical values
def convert_gender(gender):
    # Check for missing values
    if pd.isna(gender) or gender == '' or gender is None:
        return 1  # missing = 1 (default to male)
    
    # Convert to string and clean
    gender_str = str(gender).strip().lower()
    
    # Check for male
    if 'male' in gender_str or 'ছেলে' in gender_str:
        return 1
    # Check for female
    elif 'female' in gender_str or 'মেয়ে' in gender_str:
        return 2
    else:
        return 1  # anything else = 1 (default to male)

# Directly replace the A2 column
df['A2'] = df['A2'].apply(convert_gender)

# Verify the replacement
print("Unique values in A2 after replacement:")
print(df['A2'].value_counts().sort_index())
print("\nValue counts:")
print(df['A2'].value_counts())

# Check first few rows
print("\nFirst 10 rows of A2 after replacement:")
print(df['A2'].head(10))
```

```python
# Convert A3 column (religion) to numerical values
def convert_religion(religion):
    # Check for missing values
    if pd.isna(religion) or religion == '' or religion is None:
        return 0  # missing = 0
    
    # Convert to string and clean
    religion_str = str(religion).strip().lower()
    
    # Check for Islam
    if 'islam' in religion_str or 'ইসলাম' in religion_str:
        return 1
    # Check for Hindu
    elif 'hindu' in religion_str or 'হিন্দু' in religion_str:
        return 2
    # Check for Christian
    elif 'christian' in religion_str or 'christ' in religion_str or 'খ্রিস্টান' in religion_str:
        return 3
    else:
        return 0  # otherwise = 0

# Directly replace the A3 column
df['A3'] = df['A3'].apply(convert_religion)

# Verify the replacement
print("Unique values in A3 after replacement:")
print(df['A3'].value_counts().sort_index())
print("\nValue counts:")
print(df['A3'].value_counts())

# Check first few rows
print("\nFirst 10 rows of A3 after replacement:")
print(df['A3'].head(10))
```

```python
# First, let's check the original B2 column values
print("Original B2 unique values (first 20):")
print(df['B2'].head(20))
print("\nUnique values in B2:")
print(df['B2'].unique())

# Convert B2 column (time spent with family) to numerical values
def convert_family_time(time_spent):
    # Check for missing values
    if pd.isna(time_spent) or time_spent == '' or time_spent is None:
        return 1  # missing = 1 (0-3 hours category)
    
    # Convert to string and clean
    time_str = str(time_spent).strip().lower()
    
    # Try to extract numerical value (hours)
    try:
        import re
        # Look for numbers in the string
        numbers = re.findall(r'\d+', time_str)
        
        if numbers:
            # Take the first number found
            hours = float(numbers[0])
            
            # Apply the rules
            if hours <= 3:
                return 1  # 0-3 hours
            elif 4 <= hours <= 6:
                return 2  # 4-6 hours
            else:
                return 3  # more than 6 hours (otherwise)
        else:
            # Check for text-based time descriptions
            if 'less than' in time_str or 'কম' in time_str or '30' in time_str:
                return 1  # less than 1 hour falls into 0-3 category
            elif 'hour' in time_str and '1' in time_str and 'min' not in time_str:
                return 1  # 1 hour falls into 0-3 category
            elif '2' in time_str or '3' in time_str:
                return 1  # 2-3 hours
            elif '4' in time_str or '5' in time_str or '6' in time_str:
                return 2  # 4-6 hours
            elif 'more than' in time_str or 'আরো বেশি' in time_str or '7' in time_str or '8' in time_str or '9' in time_str:
                return 3  # more than 6 hours
            else:
                return 1  # default to category 1
    except:
        return 1  # default to category 1 if any error

# Apply the conversion to B2 column
df['B2'] = df['B2'].apply(convert_family_time)

# Value counts for B2 column
print("\n" + "="*50)
print("Value counts for B2 (Time spent with family):")
print(df['B2'].value_counts().sort_index())

print("\nValue counts with labels and percentages:")
value_labels = {
    1: "0-3 hours",
    2: "4-6 hours", 
    3: "More than 6 hours"
}
total = len(df)
for value in sorted(df['B2'].unique()):
    count = (df['B2'] == value).sum()
    label = value_labels.get(value, "Unknown")
    percentage = (count / total) * 100
    print(f"{label} ({value}): {count} ({percentage:.1f}%)")

# Show first 10 rows of B2 to verify
print("\nFirst 10 rows of B2 after conversion:")
print(df['B2'].head(10))
```
